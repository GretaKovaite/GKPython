```python

import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

df = pd.read_csv('../Paskaita10/pokemon.csv', encoding='utf-8')

Gen_HP = df.groupby('Generation')['HP'].mean().round(2).reset_index()

plt.plot(Gen_HP['Generation'], Gen_HP['HP'], marker='o')

plt.xlabel('Generation')
plt.ylabel('HP')
plt.title('Generation and HP ')
plt.grid(True) 
plt.show()

```

```python


type = df['Type 1'].value_counts()

ax = type.plot.bar(figsize=(10, 6), title='Count Type 1')
ax.set_xlabel('Type 1')
ax.set_ylabel('Count')  

plt.show()

```

```python

type = df['Type 1'].value_counts()

plt.bar(type.index, type.values)
plt.xlabel('Type 1')
plt.ylabel('Number of Pokemon')
plt.title('Count by type')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

```

```python

plt.scatter(df['Attack'], df['Defense'], alpha=0.5)
plt.title('Attack and Defense')
plt.grid(True)
plt.show()
```

```python

colors = ['green', 'gold']
df.value_counts("Legendary").plot.pie(
    y="Legendary",
    autopct="%1.2f%%",
    title="Legendary",
    labels=['Not Legendary', 'Legendary'], 
    ylabel="",  
    # legend=True, 
    colors=colors, 
    explode=[0, 0.1],
)
```
