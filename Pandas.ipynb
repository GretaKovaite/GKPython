```python
# Kodą rašykite čia...
import pandas as pd
import numpy as np

data = {
    'Prekė': ['Prekė A', 'Prekė B', 'Prekė C', 'Prekė D', 'Prekė E'],
    'Kaina': [100, 200, 150, 120, 180],  
}

df = pd.DataFrame(data)

df['Savikaina'] = df['Kaina'] / 1.20

df['Pelno'] = df['Kaina'] - df['Savikaina']

np.random.seed(0) 
df['Kategorija'] = np.where(np.random.rand(len(df)) > 0.5, 'Kategorija A', 'Kategorija B')

print(df)

kategoriju_skaicius = df.groupby('Kategorija').size().reset_index(name='Prekių skaičius')

print(kategoriju_skaicius)
```

```python


data = {
    'Prekė': ['Prekė A', 'Prekė B', 'Prekė C', 'Prekė D', 'Prekė E'],
    'Pardavimo kaina': [100, 200, 150, 120, 180], 
    'Parduotas kiekis': [10, 5, 15, 8, 12]  
}

df = pd.DataFrame(data)

np.random.seed(0)  
df['Kategorija'] = np.where(np.random.rand(len(df)) > 0.5, 'Kategorija A', 'Kategorija B')

print("Pradiniai duomenys:")
print(df)

rezultatai = df.groupby('Kategorija').agg(
    Pardavimo_kaina_vidurkis=('Pardavimo kaina', 'mean'),
    Parduotas_kiekis_suma=('Parduotas kiekis', 'sum')
).reset_index()

print("\nKategorijų rezultatai:")
print(rezultatai)
```
