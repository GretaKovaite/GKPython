```python
# Kodą rašykite čia...
import pandas as pd

df = pd.read_csv('pvz.csv', encoding='utf-8')
df
```

```python
# Kodą rašykite čia...
df['Vid'] = df.iloc[:, 2:].mean(axis=1).round(1)
df
```

```python
# Kodą rašykite čia...

best_stud = df.sort_values(by='Vid', ascending=False).iloc[0]

print(f'{best_stud['Vardas']} {best_stud['Pavardė']}, {best_stud['Vid']}')
```

```python
# Kodą rašykite čia...

best_stud = df.loc[df['Vid'] == df['Vid'].max()]

df.loc[best_stud.index, 'Vardas'] = 'Greta'
df.loc[best_stud.index, 'Pavardė'] = 'Kovaite'

df.head()
```

```python
# Kodą rašykite čia...

def stipendija_pagal_vidurki(vid):
    if vid >= 9.5:
        return 200
    elif vid >= 9.0:
        return 180
    elif vid >= 8.5:
        return 100
    else:
        return 0
```

```python
# Kodą rašykite čia...

df['Stipendija'] = df['Vid'].apply(stipendija_pagal_vidurki)
df
```

```python
# Kodą rašykite čia...

import matplotlib.pyplot as plt

stipendiju_sk = df['Stipendija'].value_counts()

plt.bar(stipendiju_sk.index.astype(str), stipendiju_sk.values)
plt.xlabel('Stipendija')
plt.ylabel('Studentai')
plt.title('Studentu gaunama stipendija')
plt.show()

```

```python
# Kodą rašykite čia...

vyru_sk = 8  
moteru_sk = 7 
data = {'Lytis': ['Vyrai', 'Moterys'], 'Skaicius': [vyru_sk, moteru_sk]}

lyciu_df = pd.DataFrame(data)

plt.bar(lyciu_df['Lytis'], lyciu_df['Skaicius'])
plt.xlabel('Lytis')
plt.ylabel('Studentu skaicius')
plt.title('Studentu skaicius pagal lyti')
plt.show()

```
